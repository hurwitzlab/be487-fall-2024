#### Jaccard Index (JI)

The Jaccard similarity index (sometimes called the Jaccard similarity coefficient) compares members for two sets to see which members are shared and which are distinct. Its a measure of similarity for the two sets of data, with a range from 0 to 1. The higher the value, the more similar the two populations. 

The formula to find the Index is:
JI = (the number of species found in both samples) / (the total number of species in both samples)

#### Calculate the JI

For the example above: JI = (4)/(6) = 0.67

This value tells you how similar the two sets are. 

Two sets that share all members would be a value of 1. the closer to 1, the more similarity (e.g. 0.9 is more similar than 0.89). 

If they share no members, they have a value of 0. 

The midway point  0.50  means that the two sets share half of the members.

In the literature, you may also see the Jaccard Distance, which is just the 1 minus the Jaccard coefficient (e.g., 1.0 - 0.67 = 0.33). 

We are using JI because it is easy to calculate and interpret, however, it is extremely sensitive to small samples sizes and data sets with missing observations.  Thus, in practice it is always best to use a few different indices.
